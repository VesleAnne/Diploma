```python
# Загрузим  датасет
df = pd.read_excel('/content/drive/MyDrive/МФТИ/вкр/faq.xlsx')
```

```python
# Добавим сгенерированные фразы в датасет
for phrase in operator_phrases:
    df = df.append({'Тема': np.nan, 'Вопрос': phrase, 'Схема ответа': 'Переключаю на оператора', 'Кнопка ссылки на вики': np.nan, 'Ссылка на wiki': np.nan}, ignore_index=True)

# Сохраним обновленный датасет
df.to_csv('/content/drive/MyDrive/МФТИ/вкр/обновленный_датасет.csv', index=False)
```
